## train a baseline classifier

```python
def img2feat(img, visualise=False):
    gray = color.rgb2gray(img)
#     gray = color.rgb2hsv(img)[:,:,-1]
#     gray = img
    return feature.hog(gray, 
                       visualise=visualise, 
                       transform_sqrt=True, 
                       pixels_per_cell=(8, 8), 
                       cells_per_block=(2, 2))
```

```python
np.histogram(vehicle_imgs[0], bins=32, range=(0, 256))
```

```python
# img = io.imread(nonvehicle_files[0])
img = io.imread(vehicle_files[100])
hog_feat, hog_img = img2feat(img, True)
fig, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].imshow(img)
axes[1].imshow(hog_img, cmap=plt.cm.gray)
```

```python
# img = io.imread(nonvehicle_files[0])
img = io.imread(nonvehicle_files[60])
hog_feat, hog_img = img2feat(img, True)
fig, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].imshow(img)
axes[1].imshow(hog_img, cmap=plt.cm.gray)
```

```python
%%time
vehicle_imgs = [io.imread(f) for f in vehicle_files]
nonvehicle_imgs = [io.imread(f) for f in nonvehicle_files]
```

```python
%time X = np.vstack([img2feat(img) for img in vehicle_imgs+nonvehicle_imgs])
X.shape
```

```python
y = np.stack(["vehicle"]*len(vehicle_files)+["nonvehicle"]*len(nonvehicle_files))
y.shape
```

```python
X, y = shuffle(X, y)
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3)
print(train_X.shape, test_X.shape, train_y.shape, test_y.shape)
```

```python
ss = StandardScaler()
train_X = ss.fit_transform(train_X)
test_X = ss.transform(test_X)
```

```python
%%time
pipe = RandomizedSearchCV(LinearSVC(), {'C': np.logspace(-3, 3, 10)}, cv=3, n_jobs=1)
pipe.fit(train_X, train_y)
```

```python
print(pipe.best_params_)
print(pipe.best_estimator_)
print(pipe.best_score_)
```

```python
model = pipe.best_estimator_
```

```python
model.score(test_X, test_y)
```

```python
test_yhat = model.predict(test_X)
i_mistakes = np.where(test_yhat != test_y)
print (i_mistakes)
```

## draw bounding box

```python
def draw_bbox(img, bbox):
    pt1, pt2 = bbox
    plot_img = cv2.rectangle(img.copy(), pt1, pt2, color=(0, 0, 255), thickness=2)
    return plot_img
```

```python
test_img = test_imgs[0]
bbox = ((100, 100), (164, 164))
plt.imshow(draw_bbox(test_img, bbox))
```

## pyramid of test images

```python
test_img = test_imgs[0]
for pyramid in transform.pyramid_gaussian(test_img, downscale=1.5):
    plt.figure(figsize=(4, 4))
    plt.imshow(pyramid)
    print(pyramid.shape)
    if min(pyramid.shape[:2]) <= 64:
        break
```

```python
def get_image_pyramid(image, scale=1.5, min_hw=(64, 64)):
    min_h, min_w = min_hw
    
    yield image
    size = 1
    while True:
        size /= scale
        img = imresize(image, size, )
        if img.shape[0] < min_h or img.shape[1] < min_w:
            break
        yield img
```

```python
for pyramid in get_image_pyramid(test_img, scale=1.2):
    plt.figure(figsize=(4, 4))
    plt.imshow(draw_bbox(pyramid, bbox=[(0,0),(64,64)]))
    print(pyramid.shape)
```
